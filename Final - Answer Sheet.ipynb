```python
import pandas as pd
import seaborn as sns
```

```python
#Loading three files
dfDemo = pd.read_csv("~/../smehkari/EconCourses/data/final/demographicData.csv")
dfcross = pd.read_csv("~/../smehkari/EconCourses/data/final/crossWalk.csv")
dfTax = pd.read_csv("~/../smehkari/EconCourses/data/final/taxData.csv")

#check basic info of files
dfDemo.info()
dfcross.info()
dfTax.info()
```

```python
#put all datasets together
dfdemocross = pd.merge(dfDemo,dfcross, how = "inner", on = "ID")
dfAll = pd.merge(dfdemocross,dfTax, how = "inner", on = "IDTax")
dfAll['City'] = dfAll['City'].astype('category')
dfAll.info()
```

```python
#Seaborn plotting
#Set the style
sns.set(style = 'white')

#plot the data and customize
sns.pairplot(dfAll[['Age', 
                    'Income',
                    'Wealth',
                    'City']],
             hue = 'City')
```

```python
#Open the given file
nameFile = open('listOfNames.txt','r')
#Read the whole file and get a list of names
listofnames = nameFile.read().splitlines()
#Close the file
nameFile.close()
#Create a list of edited name
editedName = []
#switch the first name and last name
for x in listofnames:
    firstlast = x.split()
    editedName.append(firstlast[1] + " " + firstlast[0])
#sort the list 
sortedName = editedName.sort()

#create a new file
editedFile = open('listOfNamesEdited.txt','w')

#split name in to a new line
streditedName = ""
for x in editedName:
    streditedName += x+'\n'

#write edited name into the file
editedFile.write(streditedName)
#close the file
editedFile.close()                   
```
