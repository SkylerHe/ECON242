1. <b> To order the legend you will have to figure out how to use the category_orders argument for px.line().</b>
2. <b> Add the following line to your code: <code>fig.update_traces(visible='legendonly')</code> to unselect all the plots to begin with.</b>
3. <b> You will need to add the following line to your code: <code> fig.update_layout(hovermode=_______) </code> with the appropriate hovermode included. Read the documentation and figure out which hover mode to use. Select multiple states from the legend to see how the hovermode works.</b>

```python
#load the data
df = pd.read_csv("~/../smehkari/EconCourses/data/Quiz8Files/Quiz8Data3.csv")
df = df[['State Name','Date','Unemployment Rate']].groupby(['Date','State Name'], as_index = False).sum()

#create a line plot
fig = px.line(df, 
              x = 'Date', 
              y = 'Unemployment Rate', 
              color = 'State Name', 
              labels = {'Date' : 'Year',
                        'State Name' : 'State'},
              title = 'Unemployment Rate over Time',
              range_x = [1980,2019],
              range_y = [0,20],
              template = 'plotly_dark')

#unselect all the plots to begin with.
fig.update_traces(visible='legendonly')

#with the appropriate hovermode included
fig.update_layout(hovermode= 'x unified')

fig.write_html('Q4.html')
```
