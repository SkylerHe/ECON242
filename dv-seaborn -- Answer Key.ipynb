<h1>Data Visualization - Seaborn - Reading Questions (Answer Key)</h1>

<b>Before we begin let us import the required modules for this question set.</b>

```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

<b>And also the datasets:</b>

```python
#
# Create the datframes for this exercise
#

# Create a dataframe with cross-county economics data
dfEcon = pd.read_csv("~/../smehkari/EconCourses/data/econdata1.csv")

# Load the gap minder dataset from plotly
import plotly.express as px
dfGapMinder = px.data.gapminder()
```

<b> Q1a. Set the seaborn style to be white and the palette to be bright: </b>

```python
sns.set(style = 'white')
sns.set_palette('bright')
```

<b> Q1b. Construct a simple scatter plot showing the relationship between percentage of population with a bachelors and higher degree and median household income:</b>

```python
sns.relplot(kind = 'scatter',
            x = 'bachelors_and_higher_perc',
            y = 'medianhhincome',
            data = dfEcon)
```

<b> Q1c. Construct a scatter plot showing the relationship between percentage of population with a bachelors and higher degree and median household income, but set the color and style of the plot to indicate the census region. Also replace the column names with properly formatted names. Play with the opacity too.</b>

```python
dfEcon2 = dfEcon.rename(columns = {'bachelors_and_higher_perc':'Bachelors and Higher Percentage', 
                                   'medianhhincome':'Median Household Income',
                                   'region_name':'Region Name'})

sns.relplot(kind = 'scatter',
            x = 'Bachelors and Higher Percentage',
            y = 'Median Household Income',
            hue = 'Region Name',
            style = 'Region Name',
            data = dfEcon2, 
            alpha = 0.4)
```

<b> Q1d. Construct 2x2 scatter plots showing the relationship between percentage of population with a bachelors and higher degree and median household income (one for each region) and set the color to indicate the census division:</b>

```python
sns.relplot(kind = 'scatter',
            x = 'bachelors_and_higher_perc',
            y = 'medianhhincome',
            hue = 'division_name',
            col = 'region_name', 
            col_wrap = 2,
            data = dfEcon)
```

<b> Q1e. Extract data for the US, Canada, and Mexico from the Gap Minder dataset and plot the life expectancy over time (Hint: This is a line plot). Differentiate the three by color, style, and size:</b>

```python
dfTemp = dfGapMinder[dfGapMinder['country'].isin(["United States", "Canada", "Mexico"])]

sns.relplot(kind = 'line',
            x = 'year',
            y = 'lifeExp',
            data = dfTemp, 
            hue = 'country', 
            style = 'country',
            size = 'country')
```

<b> Q1f. Plot the unemployment rates distribution as a violin plot by census region:</b>

```python
sns.catplot(x = 'region_name',
            y = 'unrate',
            data = dfEcon,
           kind = 'violin')
```

<b> Q1g. Plot the histogram of unemployment rates in the data. Include the kde plot:</b>

```python
sns.histplot(x = 'unrate', 
             data = dfEcon,
             kde = True)
```

<b> Q1h. Construct a scatter plot showing the relationship between percentage of population with a bachelors and higher degree and median household income and add a regression line with a 99% confidence interval: </b>

```python
sns.regplot(x = 'bachelors_and_higher_perc',
            y = 'medianhhincome',
            ci = 99,
            data = dfEcon)
```

<b> Q1i. Construct a joint plot showing the relationship between percentage of population with a bachelors and higher degree and median household income (Play around with the kind of the plot):</b>

```python
sns.jointplot(x = 'bachelors_and_higher_perc',
              y = 'medianhhincome',
              data = dfEcon, 
             kind = 'scatter')
```

<b> Q1j. Construct a pair plot showing the relationship between percentage of population with a bachelors and higher degree,  median household income, and unemployment rate by census region:</b>

```python
sns.pairplot(dfEcon[['bachelors_and_higher_perc', 
                    'medianhhincome',
                    'unrate', 
                    'region_name']],
            hue = 'region_name')
```

<b> Q1k. Reconstruct and save the plot from the previous part as a pdf file:</b>

```python
sns.pairplot(dfEcon[['bachelors_and_higher_perc', 
                    'medianhhincome',
                    'unrate', 
                    'region_name']],
            hue = 'region_name')
plt.savefig('check.pdf')
```
