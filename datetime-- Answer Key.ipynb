<br><b>Q1: Find below code that loads and cleans the employmentdatatimeseries.csv file: </b>

```python
#
# Loading in the employmentdatatimeseries.csv dataset
#
df = pd.read_csv("~/../smehkari/EconCourses/data/SplitFiles/employmentdatatimeseries.csv", na_values = ".")

# Extract the columns that we want to use
df = df[['DATE', 'UNRATE']]

# Convert Date to datetime64
df['DATE'] = df['DATE'].astype("datetime64")

# Displaying the final result (there should be no null values)
df.info()
df
```

<b>a) Augment the dataframe with two new columns: (1) the month component of the date and (2) the year component of the date.</b>

```python
df['MONTH'] = df['DATE'].dt.month
df['YEAR'] = df['DATE'].dt.year
df
```

<b>b) Construct a new dataframe that contains the average 'UNRATE' by year (explore the Y, YS, and BYS options).</b>

```python
dfNew = df[['DATE', 'UNRATE']]
dfNew = dfNew.resample(rule = 'Y', on = 'DATE').mean()
dfNew
```

<b>c) Construct a new dataframe that contains the average 'UNRATE' by quarter.</b>

```python
dfNew = df[['DATE', 'UNRATE']]
dfNew = dfNew.resample(rule = 'Q', on = 'DATE').mean()
dfNew
```

<b>d) Construct a new dataframe that contains the average 'UNRATE' at an interval of 4 months each.</b>

```python
dfNew = df[['DATE', 'UNRATE']]
dfNew = dfNew.resample(rule = '4M', on = 'DATE').mean()
dfNew
```

<b>h) Contruct a dataframe that contains the average within day price variation by month. Define the within day price variation as (Low-High)/((Low + High)/2). Note: This question requires you to do multiple things.</b>

```python
# Construct price variation measure
dfNew2 = df[['Date', 'High', 'Low']]
dfNew2['Price Var'] = 2*(dfNew2['High'] - dfNew2['Low'])/(dfNew2['High'] + dfNew2['Low'])

# Construct the new data frame
dfNew2['Month'] = dfNew2['Date'].dt.strftime('%B') 
dfNew2 = dfNew2[['Price Var', 'Month']].groupby('Month', as_index = False).mean()

# Display the results
dfNew2 = dfNew2.sort_values('Price Var')
dfNew2 = dfNew2.reset_index(drop = True)
dfNew2
```

<b> j) The overall daily change in a stock price can be broken into two components: (1) Overnight Change = Open - Previous Day Close, (2) Day Change = Close - Open. Create a new dataframe that contains the date and three new columns: Overnight Change, Day Change, and the Total Change = Close - Previous Day Close = Overnight Change + Day Change. 

```python
dfNew3 = df[['Date', 'Open', 'Close/Last']]
dfNew3 = dfNew3.sort_values('Date')
dfNew3['Prev Close'] = dfNew3['Close/Last'].shift(1)

dfNew3['Overnight Change'] = dfNew3['Open'] - dfNew3['Prev Close']
dfNew3['Day Change'] = dfNew3['Close/Last'] - dfNew3['Open']
dfNew3['Total Change'] = dfNew3['Close/Last'] - dfNew3['Prev Close']

dfNew3 = dfNew3.drop(columns = ['Open', 'Close/Last', 'Prev Close'])

dfNew3
```

<b> k) Use the dataframe above to calculate the average Overnight Change, Day Change, and Total Change. What do you learn?</b>

```python
dfNew3.mean()
```
