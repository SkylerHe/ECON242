<h1>Data Visualization -- Styling Dataframes -- Reading Questions (Answer Key)</h1>

<b>Before we begin let us import the required modules for this question set.</b>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
```

<b>And also the datasets:</b>

```python
# Create a dataframe with grades data
dfGrades = pd.DataFrame({
                        'Student': ['Pedro', 'Nicole', 'Yasmin', 'Matt', 'Amelie'],
                        'Mid 1': [28.0, 21.0, 27.0, 25.0, np.NaN],
                        'Mid 2': [26.0, 30.0, 17.0, 28.0, 29.0],
                        'Final': [39.0, 36.25, 30.0, 33.0, 39.0],
                        'Total': [93.5, 87.0, 74.0, 86.0, np.NaN],
                        'TAdj': [93.5, 87.25, 74.0, 86.0, 97.148]})

# Create a dataframe with weather data
np.random.seed(2)
dfWeather = pd.DataFrame({
                        'Day': range(1,31),
                        'Temp': np.random.randint(-20,40, size = 30)})

# Create a dataframe with macro data
dfMacro = pd.read_csv("~/../smehkari/EconCourses/data/macrodata1.csv")
```

<b> Q1a. Display the grade data with the null values highlighted in red. </b>

```python
dfGrades.style.highlight_null(null_color='red')
```

<b> Q1b. Display the grade data with the null values highlighted in red and changed to '-', the index removed, and all the data having two decimal places. </b>

```python
dfGrades.style.highlight_null(null_color='red')\
                 .format(na_rep = "-", precision = 2)\
                 .hide_index()
```

<b> Q1c. Display the grade data with the maximum and minimum values for the tests (i.e. Mid1, Mid2, and Final) highlighted in red and light blue respectively. </b>

```python
dfGrades.style.highlight_max(color = "red", 
                       axis = 0, 
                       subset = ["Mid 1", "Mid 2", "Final"]) \
              .highlight_min(color = "lightblue", 
                       axis = 0, 
                       subset = ["Mid 1", "Mid 2", "Final"])
```

<b> Q1d. Display the grade data with the background highlighted using the 'RdBu' colormap and all the data having two decimal places. </b>

```python
dfGrades.style.background_gradient(cmap='seismic')\
              .format(precision = 2)
```

<b> Q1e. Display the grade data with the text colored using the 'BuPu' colormap and all the data having two decimal places. </b>

```python
dfGrades.style.text_gradient(cmap='BuPu')\
   .format(precision = 2)
```

<b> Q1f. Display the grade data such that the total adjusted score is highlighted in blue if the score is greater than 90, highlighted in yellow if it is between 80-90, and highlighted in red if it is between 70-80. </b>

```python
dfGrades.style \
        .highlight_between(left = 90, right = 100, 
                           color = 'blue', subset = ['TAdj']) \
        .highlight_between(left = 80, right = 90, 
                           color = 'yellow', subset = ['TAdj']) \
        .highlight_between(left = 70, right = 80, 
                           color = 'red', subset = ['TAdj'])
```

<b> Q1g. Display the weather data such that the text is blue if the temperature is greater than 0 and red if it is less than 0. Also include a bar chart of the same color. </b>

```python
dfWeather.style.bar(color =['red','blue'], 
 align = 'zero', 
 axis = 0, 
 subset = ['Temp'])\
               .applymap(lambda x: f"color:red" if x < 0 
                         else f"color:blue", 
                         subset = ['Temp'])\
               .hide_index()
```

<b> Q1h. Create a correlation matrix heatmap of the macro data using the color map 'inferno'. </b>

```python
import seaborn as sns
corrMatrix = dfMacro.corr()
sns.heatmap(corrMatrix, cmap = 'inferno', annot = True)
```
