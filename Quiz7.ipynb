```python
# 
#Skyler He
#ECON242 Quiz 7 Take-Home Component
#Question 2
#

import requests
import bs4
import re
import pandas as pd

#get the webpage and convert it into HTML DOM
htmlDoc = requests.get("https://robins.richmond.edu/faculty-staff/index.html")
soup = bs4.BeautifulSoup(htmlDoc.text, 'html.parser')
#extract info from everyone's info box
divTags = soup.find_all(name = "div", attrs = {"class":"person col2-1"})
#append data into a list
firstName = []
lastName = []
department = []
email = []
phone = []

#use for loop to go through every tag objects
for divTag in divTags:
    #find the first name, last name
    firstName.append(divTag.attrs['first'])
    lastName.append(divTag.attrs['last'])
    #find the department 
    departClass = divTag.find(attrs = {'class':"department"})
    department.append(departClass.attrs['name'])
    #find the email address
    emailObject = re.compile(r'\w+@richmond.edu')
    emailClass = divTag.find(string = emailObject)
    email.append(emailClass)
    #find the phone number
    reObject = re.compile(r'\(\d{3}\)\s\d{3}-\d{4}')
    phoneClass = divTag.find(string = reObject)
    phone.append(phoneClass)

            
#convert a set of lists into a Pandas dataframe
dfDict = {"First Name":firstName,
        "Last Name":lastName,
          "Department":department,
          "Email":email,
         "Phone":phone}
df = pd.DataFrame.from_dict(dfDict)
df
```
