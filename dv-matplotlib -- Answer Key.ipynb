<h1>Data Visualization - Matplotlib - Reading Questions (Answer Key)</h1>

<b>Before we begin let us import the required modules for this question set.</b>

```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

<b>And also the datasets:</b>

```python
#
# Create the datframes for this exercise
#

# Create a dataframe with x, y1, and y2 values with the relationship y1 = x^2 and y2 = 8x^2 - x^3 + 2
df1 = pd.DataFrame({'x': range(0, 11),
                    'y1': [x**2 for x in range(0, 11)],
                    'y2': [(8*x**2 - x**3 + 2) for x in range(0, 11)]})

# Create a dataframe with x, y1, and y2 values which are randomly generated with some structure.
np.random.seed(2)
df2 = pd.DataFrame({'x': np.linspace(0, 10, 1000),
                    'y1': [math.log(x*np.random.rand() + 0.01) for x in np.linspace(0, 10, 1000)],
                    'y2': [x + x*np.random.rand() if x < 5 else math.log(x) + (10-x)*np.random.rand() for x in np.linspace(0, 10, 1000)]})

# Create a dataframe with macro data
dfEcon = pd.DataFrame({'Country': ['Niger', 'Switzerland', 'USA', 'Australia', 'Italy', 'Pakistan', 'Brazil', 'Turkey'],
                       'GDP Per Capita': [565, 75880, 68309, 54981, 43376, 1194, 15643, 8538], 
                       'Avg. Schooling': [1, 10.5, 12, 10.9, 7.2, 3.9, 4.9, 5.3]})

# Create a dataframe with macro data (and fix the date)
dfMacro = pd.read_csv("~/../smehkari/EconCourses/data/macrodata1.csv")
dfMacro['DATE'] = dfMacro['DATE'].astype('datetime64')

# Create a dataframe with Labor Force Participation Rates (Time-Series)
dfLFPR = pd.read_csv("~/../smehkari/EconCourses/data/LFPRData.csv")
dfLFPR['date'] = dfLFPR['date'].astype('datetime64')

# Create a dataframe with Labor Force (Cross-Sectional)
dfLF = pd.read_csv("~/../smehkari/EconCourses/data/LFData.csv")
dfLF['date'] = dfLF['date'].astype('datetime64')
dfLF = dfLF[dfLF['date'].isin(['2000-01-01', '2020-01-01'])]
dfLF = dfLF[['date', 'LF-16-19 Perc', 'LF-20-24 Perc', 'LF-25-54 Perc', 'LF-55+ Perc']]
```

<b> Q1a. Use the dataframe df2 to create a simple untitled scatter plot between x and y1</b>

```python
# Create empty figure
fig, ax = plt.subplots()

# Plot the data
ax.scatter(df2['x'], df2['y1'])
```

<b> Q1b. Use the dataframe df2 to create a scatter plot between x and y2. The markers should be circles, colored blue, with an opacity of 0.5. Title the figure 'x vs. y2' and set the x and y label as 'x' and 'y2'. </b>

```python
# Create empty figure
fig, ax = plt.subplots()

# Plot the data
ax.scatter(df2['x'], df2['y2'], 
c = 'blue', marker = 'o', alpha = 0.5)

# Customize the plot
ax.set_title('x vs. y2')
ax.set_xlabel('x')
ax.set_ylabel('y2')
```

<b> Q1c. Use dataframe df1 to create a line plot of x vs. y1 (red with circle markers and a solid line) and y2 (blue with square markers with a dashed line) on the same graph. Add a legend to indicate which graph is which. Add appropriate labels, a title, and grid lines.</b>

```python
# Create empty figure
fig, ax = plt.subplots()

# Plot the data
ax.plot(df1['x'], df1['y1'], c = 'red', marker = 'o', ls = '-')
ax.plot(df1['x'], df1['y2'], c = 'blue', marker = 's', ls = '--')

# Customize the plot
ax.legend(['x vs y1', 'x vs y2'])

ax.set_title('x vs. y1 & y2')
ax.set_xlabel('x')
ax.set_ylabel('y1 & y2')

ax.grid(True)

# Save the figure
fig.savefig('sample.pdf')
```

<b> Q1d. Use dataframe df1 to create a plot with two subplots (1) x vs. y1 (red with circle markers and a solid line) and (2) y2 (blue with square markers with a dashed line). The plots should be side by side. Add appropriate labels and titles to each subplot, and then also add a title for the full figure.</b>

```python
# Create empty figure
fig, ax = plt.subplots(1,2)

# Plot the data
ax[0].plot(df1['x'], df1['y1'], c = 'red', marker = 'o', ls = '-')
ax[1].plot(df1['x'], df1['y2'], c = 'blue', marker = 's', ls = '--')

# Customize the plot
ax[0].set_title('x vs. y1')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y1')

ax[1].set_title('x vs. y2')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y2')

fig.suptitle('x vs. y1 & y2')

# Save the figure
fig.savefig('sample.pdf')
```

<b> Q1e. Use the dataframe dfEcon to create a bar chart of years of schooling by country in ascending order (i.e. you will have to sort the data). The bar should have a red color and the background should be lightgrey. Do remember to add the appropriate labels and titles. Also, the figure size should be 10-by-10.</b>

```python
# Sort the values
dfEcon = dfEcon.sort_values(['Avg. Schooling'])

# Create empty figure
fig, ax = plt.subplots(figsize = (10, 10))

# Plot the data
ax.bar(dfEcon['Country'], dfEcon['Avg. Schooling'], color = 'red')

# Customize the plot
ax.set_facecolor('lightgrey')
ax.set_title('Avg. Years of Schooling by Country')
ax.set_xlabel('Country')
ax.set_ylabel('Avg. Years of Schooling')
```
