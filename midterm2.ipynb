<br><br><b> Q2 [4.5 points]: The next dataset we will be working with is the babies dataset. The filepath and descriptions of the columns can be found on blackboard. </b><br>
<b> Find below code that loads up all the files and adds a year column to this dataset. It will take a few seconds to run. </b>

```python
# Create an empty dataframe
df = pd.DataFrame()

# Loop through all the years
for year in range(1880, 2021):
    
    # Load up the file
    dfInt = pd.read_csv("~/../smehkari/EconCourses/data/babyNames/yob{}.txt".format(year), header = None)
    
    # Add column headings and a year column
    dfInt = dfInt.rename(columns = {0:"Name", 1:"Sex", 2:"Number"})
    dfInt['Year'] = year
    
    # Reorder the columns
    dfInt = dfInt[["Year", "Name", "Sex", "Number"]]
    
    # Concatenate (this is an accumulator loop)
    df = pd.concat([df, dfInt], axis = 0)
df
```

<b>a) Write code to construct a new column that contains the total number of names for each year (Hint: You will need to use the transform groupby here). Call this column "Total Names" </b>

```python
df["Total Names"] = df[['Number','Year']].groupby('Year',as_index = False).transform(np.sum)
df.reset_index(drop = True)      
```

<b>b) Write code to construct a new column that contains the fraction of babies within the year with that name (i.e. Number/Total Names). Call this new column "Frac" </b>

```python
df["Frac"] = (df['Number'] / df['Total Names']) * 100
df.reset_index(drop = True)
```

<b>c) Filter the dataset so it only includes all females with the name Elizabeth. Don't overwrite the main dataset, instead call this filtered dataset df2. </b>

```python
df2 = df[(df['Name'] == 'Elizabeth') & (df['Sex'] == 'F')]
df2.reset_index(drop = True)
```

<b>d) Use df2 to construct a line plot between "Year" (x-axis) vs. the median "Frac" (y-axis). Use the built-in plotting function in pandas </b>

```python
df2.plot(x="Year",y="Frac")
```

<b>e) Write code to construct a new column that contains the rank for each name within a year. The rank will be by the number of times a baby of a particular sex was named that name. The rank should be dense and descending. Call this column "Rank". </b></b>

```python
df['Rank'] = df[['Sex','Number','Year']].groupby(['Sex','Year'], as_index = False).rank(method = 'dense', ascending = False)   
df.reset_index(drop = True)
```

<b>f) Filter the dataset so it only includes males with a rank of 2. Don't overwrite the main dataset, instead call this filtered dataset df3.</b>

```python
df3 = df[(df['Rank'] == 2) & (df['Sex'] == 'M')].reset_index(drop = True)
df3
```

<br><br><b> Q3 [5.5 points]: The final dataset we will be working with is the iPEDS dataset. The filepath and descriptions of the columns can be found on blackboard. </b><br>
<b> Find below code that loads up all the files and adds a year column to this dataset. It also finds the list of liberal arts schools. It will take a little bit of time to run. </b>

```python
#
# Create a dataset about majors
#
# Create an empty dataframe
df = pd.DataFrame()

# Loop through all the years
for year in range(2009, 2021):
    
    # Load up the file
    dfInt = pd.read_csv("~/../smehkari/EconCourses/data/iPeds/c{}_a.csv".format(year))
    
    # Add a year column
    dfInt['YEAR'] = year
       
    # Concatenate (this is an accumulator loop)
    df = pd.concat([df, dfInt], axis = 0)

#
# Create a list of liberal art school UNITID's
#

# Load up the file
df2 = pd.read_csv("~/../smehkari/EconCourses/data/iPeds/hd2020.csv", encoding = 'latin')

# Create a list of all the schools with a carnegie classification of 31 (i.e. liberal art schools)
LASchools = list(df2[df2['CARNEGIE'] == 31]['UNITID'])
```

<b>a) There are a lot of columns in the dataset (df). We only need a subset. Overwrite df keeping only the following columns: ['YEAR', 'UNITID', 'CIPCODE', 'MAJORNUM', 'AWLEVEL', 'CTOTALT', 'CTOTALW']. This will take a few seconds, but will make the dataset more manageable. </b>

```python
df = df[['YEAR', 'UNITID', 'CIPCODE', 'MAJORNUM', 'AWLEVEL', 'CTOTALT', 'CTOTALW']]
df.reset_index(drop = True)
```

<b>b) Filter the dataset to only include information for AWLEVEL == 5 (that is only bachelors degrees). Overwrite the original dataset here. </b>

```python
df = df[df['AWLEVEL'] == 5]
df.reset_index(drop = True)
```

<b>c) Filter the dataset to only include data for the economics major (i.e. CIPCODE >= 45.0601 & CIPCODE <= 45.0699). Overwrite the original dataset here. </b>

```python
df = df[(df['CIPCODE'] >= 45.0601) & (df['CIPCODE'] <= 45.0699)]
df.reset_index(drop = True)
```

<b>d) Construct a new dataset that includes the sum of CTOTALT and CTOTALW by YEAR and UNITID. Overwrite the original dataset here.</b>

```python
df = df[['CTOTALT','CTOTALW','YEAR','UNITID']].groupby(['YEAR','UNITID'], as_index = False).sum()
df.reset_index(drop = True)
```

<b>e) Write code to construct a new column that contains the fraction of women (i.e. CTOTALW/CTOTALT). Call this new column "FRACW" </b>

```python
df["FRACW"] = df['CTOTALW']/df['CTOTALT']
df.reset_index(drop = True)
```

<b> The dataset you have from part d gives you the number of Econ majors (total and women) and fraction of women by year and school (i.e. unitid) </b>

<b>f) Construct a new dataframe dfALL, that contains the median FRACW by year for all schools</b>

```python
dfALL = df[['YEAR','FRACW']].groupby(['YEAR'], as_index = False).median()
dfALL.dropna().reset_index(drop = True)
```

<b>g) Construct a new dataframe dfLA, that contains the median FRACW by year for all liberal art schools (i.e. the unit ids in the list LASchools)</b>

```python
dfNew = pd.merge(df2,df, how = 'left', on = 'UNITID', indicator = False)
dfNew = dfNew[['UNITID','YEAR','FRACW']]
dfNew.dropna()
dfLA = dfNew[['YEAR','FRACW']].groupby(['YEAR'], as_index = False).median()
dfLA.reset_index(drop = True)
```

```python
dfLA = df[df['UNITID'] == LASchools]
dfLA
```

<b>h) Construct a new dataframe dfUR, that contains the median FRACW by year for University of Richmond (Note: UR's unit id == 233374)</b>

```python
dfUR = df[df['UNITID'] == 233374]
dfUR = dfUR[['YEAR','FRACW']].groupby(['YEAR'], as_index = False).median().reset_index(drop = True)
dfUR         
```

<b>i) Use Matplotlib to plot the FRACW for all school, liberal art schools, and University of Richmond across years. Be sure to label the x- and y-axis, title the chart, and include a legend.</b>

```python
# Create an empty figure to plot the data
fig, ax = plt.subplots(figsize = (10,10))

#Plot the FRACW for all school, liberal art schools, and University of Richmond across years.
ax.plot(dfALL['YEAR'],dfALL['FRACW'], c = 'Black', marker = '.', ls = '-')
ax.plot(dfLA['YEAR'],dfLA['FRACW'], c = 'Red', marker = 'o', ls = '--')
ax.plot(dfUR['YEAR'],dfUR['FRACW'], c = 'Blue', marker = 's', ls = ':')


#Customize the plot
ax.legend(['All schools','Liberal art schools','University of Richmond'])

#title the chart
ax.set_title('FRACW for ALL, LA, UR')

#label the x- and y- axis
ax.set_xlabel('Year')
ax.set_ylabel('FRACW')
```
